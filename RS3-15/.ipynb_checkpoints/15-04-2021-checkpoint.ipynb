## MinMax Scaler()

it will scale all the data point in between 0 to 1(or -1 to +1)

## MinMax scaler formula
# xi-min(x)
# ---------------
# max(x)-min(x)

##  Robust scaler
##   if your dataset contain more no  of outliers,then robust scaler is best technique for scaling 
##  formula
##  =========



##    xi-Q2(x)
## ----------------
## Q3(x)-Q1(x)

## Inter Quartile Range(IQR)
##  Q1=25%(Lower Quartile)
##  Q2=50%(Medium Quartile)
##  Q3=75%(Upper Quartile)
